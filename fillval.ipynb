```python
import pandas as pd
```

```python
df=pd.read_csv("data3.csv")
```

```python
df
```

```python
df.fillna(0) #fill 0 in null value
```

```python
df.fillna(2) #fill 2 in null value
```

```python
df.fillna({"Companies":"none","Profit":0 }) #(value are null) fill spacific value for spacif column Companies column none and Profic column 0
```

```python
df.fillna(method="ffill") #fill forword column value in row 
#same as pad
```

```python
df.fillna(method="bfill") #fill backword column value in row 
```

```python
df.fillna(method="ffill" ,axis=0) #fill forword value in row for particular column, forword row value fill
```

```python
df.fillna(method="ffill" ,axis=1) #fill forword value in column for particular column, forword column value fill
```

```python
df.fillna(0, limit=2) #fill top 2 row value in every column
```
