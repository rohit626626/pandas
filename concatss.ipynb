```python
import pandas as pd
```

```python
df1=pd.DataFrame({"ID":[1,2,3,4],"Name":["A","B","C","D"],"Class":[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({"ID":[5,6,7,8],"Name":["E","F","G","H"],"Class":[9,10,11,12]})
df2
```

```python
pd.concat([df1, df2]) #concat row by
```

```python
pd.concat([df1, df2],axis=1) #concat column by
```

```python
pd.concat([df1, df2], ignore_index=True) #index reset
```

```python
df1=pd.DataFrame({"ID":[1,2,3,4],"Name":["A","B","C","D"],"Class":[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({"ID":[5,6],"Name":["E","F"],"Class":[9,10]})
df2
```

```python
pd.concat([df1,df2]) #concat two unequal df row by
```

```python
pd.concat([df1,df2], axis=1) #concat two unequal df column by, value not same give nan
```

```python
pd.concat([df1,df2], axis=1, join='inner') #concat two unequal df column by, empty value row skip
```

```python
pd.concat([df1, df2],keys=['df1','df2']) #given label in df
```

```python
pd.concat([df1, df2],axis=1 ,keys=['df1','df2']) #given label in df column by
```
