```python
import pandas as pd
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],'Class':[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','C','D']})
df2
```

```python
pd.merge(df1, df2, on="ID") #merge acording id, (id is same in first and second df)(similar as sql join)
```

```python
pd.merge(df1, df2, on="ID", how='inner') #merge intercetion
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],'Class':[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,5],'Name':['A','B','C','D']})
df2
```

```python
pd.merge(df1, df2, on="ID") ##merge acording id, (id is same in first and second df), id not same value skip
```

```python
pd.merge(df1, df2, on="ID", how='inner') #merge intercetion(Default)
```

```python
pd.merge(df1, df2, on="ID", how='left') #left df complet add and right df not same given nan
```

```python
pd.merge(df1, df2, on="ID", how='right') #right df complet add and left df not same given nan
```

```python
pd.merge(df1, df2, on="ID", how='outer') #right and left df complet add, not same given nan
```

```python
pd.merge(df1, df2, on="ID", how='outer', indicator=True) #right and left df complet add, not same given nan, indicator given they value this df
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],'Class':[9,10,11,12]})
df1
```
