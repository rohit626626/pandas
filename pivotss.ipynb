```python
import pandas as pd
```

```python
df=pd.read_csv("data7.csv")
df
```

```python
df.pivot_table(index='Genre') #Genre column index and given mean value
```

```python
df.pivot_table(index='Genre', columns='Year')#given value(mean) year by
```

```python
df.pivot_table(index='Genre', columns='Year', aggfunc='sum')#given value(sum) year by
```

```python
df.pivot_table(index='Genre', columns='Year', fill_value=0)#fill nan value to 0
```

```python
df.pivot_table(index='Genre', columns='Year', aggfunc='sum',fill_value=0, margins=True) #given sum row by and column by
```
