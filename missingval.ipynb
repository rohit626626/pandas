```python
df=pd.read_csv("data1.csv")
```

```python
df.isnull() #given true value is nan, null, empty otherwise false table 
#this work is series
```

```python
df.isnull().sum() #given column by value is null 
```

```python
df.isnull().sum().sum() #given total value is null
```

```python
df.notnull() #given false value is nan, null, empty otherwise true table 
```

```python
df.notnull().sum() #given column by value is not null
```

```python
df.notnull().sum().sum() #given total value is not null
```

```python
df3=pd.read_csv("data3.csv")
df3
```

```python
df3.dropna() # one value are null in row, delete row
```

```python
df3.dropna(axis=1) # one value are null in column, delete column
```

```python
df3.dropna(how="any") # one value are null in row, delete row by default any
```

```python
df3.dropna(axis=1,how="any") # one value are null in column, delete column by default any
```

```python
df.dropna(how="all") #complet row is empty then row deleted
```

```python
df.dropna(thresh=8) # less 8 not null value in row, row deleted
```

```python
df.dropna(subset=["Companies"]) #companies column null value, they row deleted 
```
