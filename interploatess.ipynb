```python
df=pd.read_csv("data6.csv",parse_dates=['Date'], index_col="Date")
```

```python
df.interpolate(method='polynomial',order=1) #fill value polynomial by acording to order 
```

```python
df.interpolate(method='spline',order=1)
```

```python
df.interpolate(limit=1) #fill 1 value in gap
```

```python
df.interpolate(limit=1, limit_direction='backward') #fill 1 value in gap to backward(bottom)
```

```python
df.interpolate(limit=1, limit_direction='both') #fill 1 value in gap to 1 top and 1 bottom
```

```python
df.interpolate(limit_area='inside') #fill value inside
```

```python
df.interpolate(limit_area='outside') #fill value outside
```
